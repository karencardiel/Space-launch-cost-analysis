##The data

**Cost of space launches to low Earth orbit**

The dataset was obtained from [Our World in Data](https://ourworldindata.org/grapher/cost-space-launches-low-earth-orbit) by clicking on download > Full data (CSV).

It apparently has data of each inital flight of a rocket and its cost to carry a kg to space.

Variables:

*  `Entity` contains the model of the rocket
*  `Year` its the first launch year
*  `cost_per_kg` is the cost to carry a kg to space
*  `launch_class` is whether is a heavy, medium or small rocket











```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from google.colab import drive
drive.mount("/content/drive/")
```

#**Data table**

```python
launches=pd.read_csv("/content/drive/MyDrive/datasets/cost-space-launches-low-earth-orbit.csv")
launches.head()
```

#**Histrogram charts**

```python
sns.histplot(launches['Year'], color='#7C00FE', alpha=0.6)

plt.title('Número de primeros lanzamientos con éxito por año')
plt.xlabel('Año')
plt.ylabel('Cantidad de lanzamientos')

plt.grid(True, axis='y', color='gray', linestyle='--', linewidth=0.5)
plt.show()
```

```python
sns.histplot(data=launches, x='launch_class', color='#009FBD', alpha=0.6)

plt.title('Cantidad de lanzamientos por clase')
plt.xlabel('Clase de lanzamiento')
plt.ylabel('Cantidad de lanzamientos')

plt.grid(True, axis='y', color='gray', linestyle='--', linewidth=0.5)
plt.show()
```

```python
sns.histplot(data=launches, x='cost_per_kg', color='#FF6969', alpha=0.6)

plt.title('Costo por kilogramo por cantidad de lanzamientos')
plt.xlabel('Costo por kilogramo')
plt.ylabel('Cantidad de lanzaminetos')

plt.grid(True, axis='y', color='gray', linestyle='--', linewidth=0.5)
plt.show()
```

#**Distribution chart**

```python
colors=['#7C00FE','#E90074', '#009FBD']

sns.scatterplot(x='Year', y='cost_per_kg', data=launches, hue='launch_class', palette=colors)

plt.grid(True, color='gray', linestyle='--', linewidth=0.5)

plt.title('Costo de lanzamiento por kilogramo de carga por año')
plt.legend(title='Payload size', title_fontsize='10')
plt.xlabel('Año')
plt.ylabel('Costo de lanzamiento por kg de carga')

plt.show()
```
